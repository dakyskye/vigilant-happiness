# 1

```python
import pandas as pd
```

```python
df = pd.read_csv("Earthquakes_database.csv")
df
```

```python
df.drop(columns='Source', inplace=True)
df
```

```python
df['Type'] = df['Type'].replace('Earthquake', 'Yay! An Earthquake')
df
```

```python
df = df.fillna("come on, not a NA anymore")
df
```

# 2

```python
bins = [0, 100, 200, float('inf')]
labels = ['Shallow', 'Medium', 'Deep']

df['Category'] = pd.cut(df['Depth'], bins=bins, labels=labels, right=False)
```

```python
df
```

```python
df['Category']
```

# 4

```python
import numpy as np
import matplotlib.pyplot as plt
```

```python
# Depth histogram
plt.figure(figsize=(10, 6))
plt.hist(df['Depth'], bins=10, color='blue', alpha=0.7)
plt.title('Histogram of Depth')
plt.xlabel('Depth')
plt.ylabel('Frequency')
plt.grid(True)
plt.show()
```

```python
plt.figure(figsize=(10, 6))
df['Type'].value_counts().plot(kind='bar', color='purple', alpha=0.7)
plt.title('Count of Events by Type')
plt.xlabel('Type')
plt.ylabel('Count')
plt.grid(True)
plt.show()
```

```python
plt.figure(figsize=(8, 8))
df['Category'].value_counts().plot(kind='pie', autopct='%1.1f%%', colors=['gold', 'lightblue', 'lightgreen'])
plt.title('Distribution of Categories')
plt.ylabel('')  # Hide the y-label
plt.show()
```

```python
plt.figure(figsize=(10, 6))
plt.plot(df['Latitude'], df['Longitude'], marker='x', linestyle='-', color='red')
plt.title('Latitude vs Longitude')
plt.xlabel('Latitude')
plt.ylabel('Longitude')
plt.grid(True)
plt.show()
# this is funny :DDDDD
```

```python
df['Rolling_Mean_Magnitude'] = df['Magnitude'].rolling(window=3).mean()

print("DataFrame with Rolling Mean (Window=3 Days) for Magnitude:")
print(df)
```

```python
# Shift the Depth values by one period (day in this case)
df['Shifted_Depth'] = df['Depth'].shift(1)

print("DataFrame with Shifted Depth Values by One Day:")
print(df)
```
